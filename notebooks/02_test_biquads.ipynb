```python
import torch
import torchaudio
from dasp_pytorch import ParametricEQ

# Load test audio
audio, sr = torchaudio.load('data/raw/unmastered/song_001.wav')

# Create differentiable EQ
eq = ParametricEQ(sample_rate=sr, num_bands=5)

# Define EQ parameters
frequencies = torch.tensor([[100.0, 500.0, 2000.0, 5000.0, 10000.0]])
gains = torch.tensor([[2.0, -1.0, 3.0, 1.0, -2.0]])  # dB
q_factors = torch.tensor([[1.0, 1.5, 2.0, 1.0, 1.0]])

# Apply EQ
with torch.no_grad():
    output = eq(audio.unsqueeze(0), frequencies, gains, q_factors)

# Save and listen
torchaudio.save('test_eq_output.wav', output.squeeze(0), sr)

# Visualize frequency response
import matplotlib.pyplot as plt
import numpy as np

# Compute frequency response
freqs = np.logspace(np.log10(20), np.log10(20000), 1000)
# TODO: Plot EQ curve
```
